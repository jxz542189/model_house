```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
SIZE = 1000
NOISE_RANGE = 0.5
TREND = 200. # increase the value to get stationary series
OSCILIATION = 100. # decrease to get more peaks
np.random.seed = 19831006

x = np.array(range(SIZE))
#noise = np.random.uniform(-1 * NOISE_RANGE, NOISE_RANGE, SIZE)
noise = np.random.normal(0, NOISE_RANGE, SIZE )
y = np.sin(np.pi * x / OSCILIATION) + (x / TREND + noise)

print(len(y))
```

```python
plt.close('all')
plt.figure(figsize=(20, 10))

plt.subplot(1,1,1)
plt.title("Time Series Data")
plt.plot(x,y)
plt.xlabel("Time Index")
plt.ylabel("Value")
```

```python
x_learn = x[:700]
y_learn = y[x_learn]

x_test = x[700:]
y_test = y[x_test]
```

```python
plt.close('all')
plt.figure(figsize=(20, 8))

plt.subplot(1,2,1)
plt.title("Train Data")
plt.plot(x_learn,y_learn)
plt.xlabel("Time Index")
plt.ylabel("Value")

plt.subplot(1,2,2)
plt.title("Test Data")
plt.plot(x_test, y_test)
plt.xlabel("Time Index")
plt.ylabel("Value")
```

```python
train_df = pd.DataFrame({"time_index":x_learn, "value":y_learn})
print(len(train_df))
train_df.head()
```

```python
test_df = pd.DataFrame({"time_index":x_test, "value":y_test})
print(len(test_df))
test_df.head()
```

```python
train_df.to_csv(path_or_buf="data/train-data.csv", header=True, index=False)
test_df.to_csv(path_or_buf="data/test-data.csv", header=True, index=False)
```
